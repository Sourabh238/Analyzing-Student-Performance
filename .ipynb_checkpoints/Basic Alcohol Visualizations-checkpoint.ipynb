```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
sns.set_color_codes("pastel")
%matplotlib inline
```

```python
# This will be our transformed dataframes
mathTransformed = pd.read_csv('math_transformed.csv')
portTransformed = pd.read_csv('port_transformed.csv')
```

```python
mathTransformed['Avg'] = (mathTransformed['G3'] + mathTransformed['G1'] + mathTransformed['G2']) / 3.0
portTransformed['Avg'] = (portTransformed['G3'] + portTransformed['G1'] + portTransformed['G2']) / 3.0
```

```python
ax = sns.regplot(x="Walc", y="Avg", data=mathTransformed, marker = '+')
ax.grid(False)
ax.set_title("Portuguese")
```

```python
portS = ['Portuguese' for i in range(649)]
mathS = ['Math' for i in range(395)]
portAvg = list(portTransformed['Avg'])
mathAvg = list(mathTransformed['Avg'])
subject = portS + mathS
all_grades = portAvg + mathAvg
daily_port = list(portTransformed['Dalc']) 
daily_math = list(mathTransformed['Dalc'])
weekend_port = list(portTransformed['Walc'])
daily = list(portTransformed['Dalc']) + list(mathTransformed['Dalc'])
weekend_math = list(mathTransformed['Walc'])
weekend = list(portTransformed['Walc']) + list(mathTransformed['Walc'])
```

```python
d = {'Subject':portS, 'Average Grade': portAvg, 'Weekday Alcohol Consumption': daily_port}
daily_alc_port = pd.DataFrame(data=d)
d1 = {'Subject':mathS, 'Average Grade': mathAvg, 'Weekday Alcohol Consumption': daily_math}
daily_math_alc = pd.DataFrame(data=d1)
```

```python
d = {'Subject':portS, 'Average Grade': portAvg, 'Weekday Alcohol Consumption': weekend_port}
weekend_alc_port = pd.DataFrame(data=d)
d1 = {'Subject':mathS, 'Average Grade': mathAvg, 'Weekday Alcohol Consumption': weekend_math}
weekend_alc_port = pd.DataFrame(data=d1)
```

```python
d = {'Subject':portS, 'Average Grade': portAvg, 'Weekday Alcohol Consumption': daily}
daily_alc = pd.DataFrame(data=d)
d1 = {'Subject':subject, 'Average Grade': all_grades, 'Weekend Alcohol Consumption': weekend}
weekend_alc = pd.DataFrame(data=d1)
```

```python
ax = sns.stripplot(x = 'Weekday Alcohol Consumption', y = 'Average Grade', zorder=0, data=daily_alc_port)
ax = sns.boxplot(x = 'Weekday Alcohol Consumption', y = 'Average Grade', data=daily_alc_port, showcaps=False, boxprops={'facecolor':'None'}, showfliers=False,whiskerprops={'linewidth':0})
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2])
ax.grid(False)
```

```python
#ax = sns.violinplot(x = 'Model', y = 'Errors', data=df_port)

ax = sns.stripplot(x = 'Weekday Alcohol Consumption', y = 'Average Grade', jitter = 0.0, zorder=0, hue='Subject', data=daily_alc, palette="Set2", split = True)
ax = sns.boxplot(x = 'Weekday Alcohol Consumption', y = 'Average Grade', hue='Subject', data=daily_alc, palette = "Set2", showcaps=False, boxprops={'facecolor':'None'}, showfliers=False,whiskerprops={'linewidth':0})
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2])
ax.grid(False)
ax.figure.savefig('Weekday_Alcohol.png')
```

```python
#ax = sns.violinplot(x = 'Model', y = 'Errors', data=df_port)


ax = sns.stripplot(x = 'Weekend Alcohol Consumption', y = 'Average Grade', jitter = 0.0, zorder=0, hue='Subject', data=weekend_alc, palette="Set2", split = True)
ax = sns.boxplot(x = 'Weekend Alcohol Consumption', y = 'Average Grade', hue='Subject', data=weekend_alc, palette = "Set2", showcaps=False, boxprops={'facecolor':'None'}, showfliers=False,whiskerprops={'linewidth':0})
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles[:2], labels[:2])
ax.grid(False)
ax.figure.savefig('Weekend_Alcohol.png')
```
