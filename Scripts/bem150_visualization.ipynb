```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # visualize
import matplotlib.pyplot as plt
%matplotlib inline

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

# Any results you write to the current directory are saved as output.

data = pd.read_csv('math_transformed.csv')

data.columns

data['Dalc'] = data['Dalc'] + data['Walc']

ave = sum(data.G1)/float(len(data))
data['ave_line'] = ave
data['average'] = ['above average' if i > ave else 'under average' for i in data.G1]
ax = plt.axes()
sns.swarmplot(x='Dalc', y = 'G1', hue = 'average',data= data,palette={'above average':'lime', 'under average': 'red'})
ax.set(xlabel='Weekly Alcohol Consumption', ylabel='First Semester Grade')
ax.set_title('First Semester Grade vs Weekly Alcohol Intake for Math Students')
plt.savefig('above_below_math.png')
```
